```python
# 불필요한 행/열 삭제
minsu = minsu.drop([0, 1]).drop(columns=['Unnamed: 0']).reset_index(drop=True)
```

```python
# 필요한 칼럼만 남기기
col = ['터미널코드', '선석코드', '터미널모선항차', '접안예정일시', '출항예정일시', '양하완료수량', '양하잔여수량', '양하합계수량', '적하완료수량', '적하잔여수량', '적하합계수량',]
minsu = minsu[col]
```

```python
# 데이터 복사
minsu1 = minsu.copy()
```

```python
# 데이터 분리 (날짜 / 시각)
minsu1[['접안예정일시_날짜', '접안예정일시_시각']] = minsu1['접안예정일시'].str.split(' ', n=1, expand=True)
minsu1.drop(columns=['접안예정일시'], inplace=True)

minsu1[['출항예정일시_날짜', '출항예정일시_시각']] = minsu1['출항예정일시'].str.split(' ', n=1, expand=True)
minsu1.drop(columns=['출항예정일시'], inplace=True)
```

```python
# '~_시각' 전처리 함수

def convert_to_24hr(time_str):
    # 데이터가 문자열이 아니거나 결측치인 경우 그대로 반환
    if not isinstance(time_str, str) or pd.isna(time_str):
        return time_str
    
    # 시간 문자열을 공백을 기준으로 분할
    time_parts = time_str.split()
    
    # 분할된 문자열이 유효한지 확인
    if len(time_parts) != 2:
        return time_str
    
    # 오전/오후와 시간을 분리합니다.
    am_pm = time_parts[0].strip()
    time_digits = time_parts[1].strip().split(':')
    
    # 시간과 분을 추출하고 필요에 따라 0을 추가합니다.
    hour = time_digits[0].zfill(2)
    minute = time_digits[1].zfill(2)
    second = time_digits[2].zfill(2) if len(time_digits) > 2 else '00'

    # 12시간제를 24시간제로 변환합니다.
    if am_pm == '오후':
        hour = str(int(hour) + 12)
        if hour == '24':
            hour = '12'
    elif am_pm == '오전' and hour == '12':
        hour = '00'

    return f"{hour}:{minute}:{second}"
```

```python
# '~시각' 결측치 '00:00:00' 대체
minsu1['접안예정일시_시각'] = minsu1['접안예정일시_시각'].apply(convert_to_24hr).fillna('00:00:00')
minsu1['출항예정일시_시각'] = minsu1['출항예정일시_시각'].apply(convert_to_24hr).fillna('00:00:00')
```

```python
# '~_날짜' 결측치 '0000_00_00' 대체
minsu1['접안예정일시_날짜'] = minsu1['접안예정일시_날짜'].fillna('0000_00_00')
minsu1['출항예정일시_날짜'] = minsu1['출항예정일시_날짜'].fillna('0000_00_00')
```

```python
# '~수량' 결측치 0으로 대체

minsu1['양하완료수량'].fillna(0, inplace=True)
minsu1['양하잔여수량'].fillna(0, inplace=True)
minsu1['양하합계수량'].fillna(0, inplace=True)
minsu1['적하완료수량'].fillna(0, inplace=True)
minsu1['적하잔여수량'].fillna(0, inplace=True)
minsu1['적하합계수량'].fillna(0, inplace=True)
```
