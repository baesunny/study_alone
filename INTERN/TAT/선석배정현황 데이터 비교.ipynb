# 3. Data Preprocessing

## 3-1 ) minsu

```python
minsu = pd.read_csv('정민수_선석스케줄.csv')
minsu.head()
```

```python
# 불필요한 행/열 삭제
minsu = minsu.drop([0]).drop(columns=['Unnamed: 0']).reset_index(drop=True)
```

```python
# 첫 번째 행을 칼럼명으로 설정
minsu.columns = minsu.iloc[0]
minsu = minsu.drop(0)
```

```python
# 필요한 칼럼만 남기기
col = ['터미널코드', '모선항차', '선사항차', '선사', '항로', '접안방향', '접안예정일시', '출항예정일시', '반입마감일시', '선석코드', '양하수량', '적하수량', '이적수량']
minsu = minsu[col]
```

                                                -

```python
# 데이터 복사
minsu1 = minsu.copy()
```

```python
# 결측치 처리 전
minsu1.isna().sum()
```

```python
# '~수량' 결측치 0으로 대체
minsu1['양하수량'].fillna(0, inplace=True)
minsu1['적하수량'].fillna(0, inplace=True)
minsu1['이적수량'].fillna(0, inplace=True)

# '~일시' 결측치 '0000-00-00 00:00:00' 대체
minsu1['반입마감일시'].fillna('0000-00-00 00:00:00', inplace=True)

# '선사' 결측치 'FALSE' 대체
minsu1['선사'].fillna('FALSE', inplace=True)

# '항로' 결측치 'FALSE' 대체
minsu1['항로'].fillna('FALSE', inplace=True)
```

```python
# 최종 결측치 0개
minsu1.isna().sum()
```
